# Seaborn

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
fmri.tail(50)
```

```python
sns.lineplot(x = "timepoint",y = "signal",data= fmri,hue="event",style="event",markers=True)
plt.show()
```

# Bar Plot

```python
sns.set(style="whitegrid")
pokemon= pd.read_csv('pokemon.csv')
sns.barplot(x="is_legendary",y="speed",hue="generation",data=pokemon)
plt.show()
```

# ScatterPlot

```python
Iris = pd.read_csv('Iris.csv')
Iris.head()
sns.scatterplot(x='SepalLengthCm',y="PetalLengthCm",hue="Species",style="Species",data=Iris)
plt.show()
```
