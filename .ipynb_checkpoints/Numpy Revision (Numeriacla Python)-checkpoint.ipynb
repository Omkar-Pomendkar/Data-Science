```python
# Joining Numpy Array
# vstack() - Vertical Stacking 
n10 = np.array([1,2,3,4,5])
n11 = np.array([6,7,8,9,10])

np.vstack((n10,n11))
```

```python
# Joining Numpy Array
# hstack() - Horizontal Stacking 
n12 = np.array([10,20,30,40,50])
n13 = np.array([60,70,80,90,100])

np.hstack((n12,n13))
```

```python
# Joining Numpy Array
# cstack() - Column Stacking 
n14 = np.array([10,20,30,40,50])
n15 = np.array([60,70,80,90,100])

np.column_stack((n14,n15))
```

```python
# Numpy insertion & Difference
n14 = np.array([10,20,30,40,50])
n15 = np.array([60,20,30,70,80,90])

```

```python
np.intersect1d(n14,n15) # Common Elements
```

```python
np.setdiff1d(n15,n14) # shows differnce between two list
```

```python
# Adding Numpy Arrays
 
n16 = np.array([12,34])
n17 = np.array([23,56])

np.sum([n16,n17])
```

```python
# NUmpy Basic Addition
new1 = np.array([10,20,30,40])

new1=new1+1
new1
```

```python
# NUmpy Basic Subrtaction
new2 = np.array([10,20,30,40])

new2=new2-3
new2
```

```python
# NUmpy Basic MUltiplication
new3 = np.array([10,20,30,40])

new3=new2*3
new3
```

```python
# NUmpy Basic division
new4 = np.array([10,20,30,40])

new4=new4/3
new4
```

```python
# Mean
no1 = np.array([12,23,45,34,56,78,90])

np.mean(no1)

```

```python
#Matrix Multiplication

a1 = np.array([[1,3,0],[5,6,7],[12,9,10]])

a2 = np.array([[1,2,5],[2,1,7],[6,7,10]])

a1.dot(a2)
a2.dot(a1)
```

```python
# Saving Numpy
sa = np.array([[1,2,5],[2,1,7],[6,7,10]])
np.save('my_numpy_saved',sa)
```
